## Punto 1: Estrategia de partición en entrenamiento y test para el pandas dataframe (car_evaluation)

La variable Class, está concentrada en el nivel Unacc 70%. Para evitar que las particiones tengan una distribución diferente, se necesita hacer un muestreo aleatorio estratificado por esta variable.

Se procede a diseñar un programa que extraiga en muestreo aleatorio estratificado por class el 30% de la base original y la tome como "TEST" y el 70% restante la tome como "TRAIN", el proceso de estratificación, asegura que el muestreo en ambas bases contengan la misma estructura en la depediente.

```python

from sklearn.model_selection import train_test_split

train, test = train_test_split(car_evaluation, test_size=0.3, stratify=car_evaluation['class'], random_state=42)

print("Distribución en el conjunto de entrenamiento:")
print(train['class'].value_counts(normalize=True))

print("\nDistribución en el conjunto de prueba:")
print(test['class'].value_counts(normalize=True))
```

## Punto 2: Propuesta de modelos y comparación de desempeño

Se proponen cinco rutas diferentes: K-Nearest Neighbors,Decision Tree, Random Forest, Support Vector Classifier, Multinomial Logistic Regression. A partir de la tabla de clasificación, se procede a comparar el desempeño predictivo de las cinco propuestas comparando desempeño en Entrenamiento y Test.

Los resultados mostrarán que el modelo de arboles de clasificación es el modelo que en TEST presenta menor cantidad de falsos positivos, menor cantidad de falsos negativos y mayor proporción de clasificación correcta.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix


X_train = pd.get_dummies(train.drop('class', axis=1)).values
y_train = train['class'].values
X_test = pd.get_dummies(test.drop('class', axis=1)).values
y_test = test['class'].values

# definición de modelos candidatos
models = {
    "K-Nearest Neighbors": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Support Vector Classifier": SVC(random_state=42),
    "Multinomial Logistic Regression": LogisticRegression(multi_class='multinomial', solver='lbfgs', random_state=42)
}


results = []

# Definición de matrices de confusión
def calculate_metrics(y_true, y_pred):
    confusion = confusion_matrix(y_true, y_pred, labels=["unacc", "acc", "good", "vgood"])
    tn = confusion[0, 0]
    fp = confusion[0, 1:].sum()
    fn = confusion[1:, 0].sum()
    tp = confusion[1:, 1:].sum()
    total = confusion.sum()
    correct_classification = (tp + tn) / total * 100
    fp_percentage = fp / total * 100
    fn_percentage = fn / total * 100
    return correct_classification, fp_percentage, fn_percentage

# Entrenamiento y evaluación de modelos
for name, model in models.items():
    # Entrenamiento
    model.fit(X_train, y_train)
    
    # Evaluación en entrenamiento
    y_train_pred = model.predict(X_train)
    correct_classification_train, fp_train, fn_train = calculate_metrics(y_train, y_train_pred)
    
    # Evaluación en test
    y_test_pred = model.predict(X_test)
    correct_classification_test, fp_test, fn_test = calculate_metrics(y_test, y_test_pred)
    
    # generación de diccionarios
    results.append({
        "Algoritmo": name,
        "Train - Clasificación Correcta (%)": correct_classification_train,
        "Train - Falso Positivo (%)": fp_train,
        "Train - Falso Negativo (%)": fn_train,
        "Test - Clasificación Correcta (%)": correct_classification_test,
        "Test - Falso Positivo (%)": fp_test,
        "Test - Falso Negativo (%)": fn_test
    })

# resultados a dataframe
results_df = pd.DataFrame(results)

# Mostrar la tabla comparativa
print("Tabla Comparativa de Resultados para TRAIN y TEST (Frecuencia Relativa %):")
print(results_df)
```

## Punto 3: Mejoramiento del modelo

Teniendo en cuenta que el modelo con mejor desempeño por tener la mejor tasa de clasificación correcta y al mismo tiempo menor tasa de falsos positivos fue el arbol de decisión, se planifica probar diferentes parámetros de entrenamiento para comparar la mejor estrategia de arbol de decisión. En este caso se planifican algoritmos de crecimiento y poda diversos iterando hasta encontrar una combinación eficiente comparando hasta encontrar una solución estable.

Se escoge el modelo que en TEST obtuvo los mejores resultados posibles: Clasificación correcta:98,8%, Falsos positivos: 0,96 Falsos negativos: 0,19. Esta configuración está compuesta por un algoritmo de crecimiento y poda con los siguientes parámetros:

Criterio de división (criterion): gini. Este criterio utiliza el índice de Gini para evaluar la calidad de las divisiones en cada nodo del árbol. El índice de Gini mide la impureza de una partición, favoreciendo divisiones que maximizan la pureza de las clases dentro de los nodos resultantes, es decir, aquellas en las que la mayoría de las observaciones pertenecen a una misma clase.

Profundidad máxima (max_depth): 10. La profundidad del árbol se limita a un máximo de 10 niveles, lo cual ayuda a controlar la complejidad del modelo. Esta restricción reduce el riesgo de sobreajuste al limitar la capacidad del modelo de aprender detalles demasiado específicos de los datos de entrenamiento, mejorando así su capacidad de generalización en datos nuevos.

Mínimo de muestras por división (min_samples_split): 2. Se requiere un mínimo de dos muestras para intentar dividir un nodo. Este valor permite al modelo realizar todas las divisiones posibles, lo cual maximiza la flexibilidad en el crecimiento del árbol dentro de la profundidad permitida.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix

# Definir diferentes configuraciones de árboles de decisión
tree_configs = [
    {"criterion": "gini", "max_depth": None, "min_samples_split": 2},
    {"criterion": "gini", "max_depth": 5, "min_samples_split": 2},
    {"criterion": "gini", "max_depth": 10, "min_samples_split": 2},
    {"criterion": "entropy", "max_depth": None, "min_samples_split": 2},
    {"criterion": "entropy", "max_depth": 5, "min_samples_split": 2},
    {"criterion": "entropy", "max_depth": 10, "min_samples_split": 2},
    {"criterion": "gini", "max_depth": None, "min_samples_split": 10},
    {"criterion": "entropy", "max_depth": None, "min_samples_split": 10},
]

# Lista para almacenar los resultados
results = []

# Función para calcular métricas a partir de la matriz de confusión
def calculate_metrics(y_true, y_pred):
    confusion = confusion_matrix(y_true, y_pred, labels=["unacc", "acc", "good", "vgood"])
    tn = confusion[0, 0]
    fp = confusion[0, 1:].sum()
    fn = confusion[1:, 0].sum()
    tp = confusion[1:, 1:].sum()
    total = confusion.sum()
    correct_classification = (tp + tn) / total * 100
    fp_percentage = fp / total * 100
    fn_percentage = fn / total * 100
    return correct_classification, fp_percentage, fn_percentage

# Entrenar y evaluar cada configuración de árbol de decisión
for config in tree_configs:
    # Crear el modelo con los hiperparámetros actuales
    model = DecisionTreeClassifier(criterion=config["criterion"], 
                                   max_depth=config["max_depth"], 
                                   min_samples_split=config["min_samples_split"], 
                                   random_state=42)
    
    # Entrenar el modelo en el conjunto TRAIN
    model.fit(X_train, y_train)
    
    # Evaluar en TRAIN
    y_train_pred = model.predict(X_train)
    correct_classification_train, fp_train, fn_train = calculate_metrics(y_train, y_train_pred)
    
    # Evaluar en TEST
    y_test_pred = model.predict(X_test)
    correct_classification_test, fp_test, fn_test = calculate_metrics(y_test, y_test_pred)
    
    # Almacenar los resultados en un diccionario
    results.append({
        "Configuración": f"criterion={config['criterion']}, max_depth={config['max_depth']}, min_samples_split={config['min_samples_split']}",
        "Train - Clasificación Correcta (%)": correct_classification_train,
        "Train - Falso Positivo (%)": fp_train,
        "Train - Falso Negativo (%)": fn_train,
        "Test - Clasificación Correcta (%)": correct_classification_test,
        "Test - Falso Positivo (%)": fp_test,
        "Test - Falso Negativo (%)": fn_test
    })

# Convertir los resultados a un DataFrame
results_df = pd.DataFrame(results)

# Mostrar la tabla comparativa
print("Tabla Comparativa de Diferentes Configuraciones de Árboles de Decisión para TRAIN y TEST (Frecuencia Relativa %):")
print(results_df)
```

## Punto 4: Evaluación en profundidad del algoritmo de arbol con mejor desempeño

Se procede a analizar en profundidad el modelo, se observa que las altas clasificaciones se deben a que en los nodos terminales del arbol quedaban muy pocos casos.

Se agrega una condición para que mínimo en cada nodo terminal participen 45 casos. De igual forma se imprime el arbol y se prepara todo para analizar nuevamente bondad de ajuste con esta regla de poda.

La recla de poda, reduce la capacidad de predicción correcta a 88,6% sin embargo reduce la posibilidad de sobre ajuste dado que los anteriores algoritmos lograban predicciones altas en nodos con baja muestra.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import confusion_matrix, classification_report
import matplotlib.pyplot as plt

# Dividir en conjuntos de entrenamiento y prueba con muestreo estratificado
train, test = train_test_split(car_evaluation, test_size=0.3, stratify=car_evaluation['class'], random_state=42)

# Separar características (X) y la variable dependiente (y) y codificar variables categóricas
X_train = pd.get_dummies(train.drop('class', axis=1))  # Aplicar pd.get_dummies para codificación
y_train = train['class']
X_test = pd.get_dummies(test.drop('class', axis=1))
y_test = test['class']

# Configuración del modelo de árbol de decisión con restricción de casos mínimos en nodo terminal
model = DecisionTreeClassifier(criterion="gini", max_depth=10, min_samples_split=2, min_samples_leaf=45, random_state=42)

# Entrenar el modelo en el conjunto TRAIN
model.fit(X_train, y_train)

# Función para calcular métricas y generar tablas de clasificación
def calculate_metrics(y_true, y_pred):
    confusion = confusion_matrix(y_true, y_pred, labels=["unacc", "acc", "good", "vgood"])
    tn = confusion[0, 0]
    fp = confusion[0, 1:].sum()
    fn = confusion[1:, 0].sum()
    tp = confusion[1:, 1:].sum()
    total = confusion.sum()
    correct_classification = (tp + tn) / total * 100
    fp_percentage = fp / total * 100
    fn_percentage = fn / total * 100
    return correct_classification, fp_percentage, fn_percentage

# Evaluación en el conjunto TRAIN
y_train_pred = model.predict(X_train)
correct_classification_train, fp_train, fn_train = calculate_metrics(y_train, y_train_pred)

# Evaluación en el conjunto TEST
y_test_pred = model.predict(X_test)
correct_classification_test, fp_test, fn_test = calculate_metrics(y_test, y_test_pred)

# Generar la tabla de evaluación
results = {
    "Conjunto": ["Train", "Test"],
    "Clasificación Correcta (%)": [correct_classification_train, correct_classification_test],
    "Falso Positivo (%)": [fp_train, fp_test],
    "Falso Negativo (%)": [fn_train, fn_test]
}
results_df = pd.DataFrame(results)

# Mostrar la tabla comparativa
print("Tabla Comparativa para el Modelo de Árbol de Decisión (Frecuencia Relativa %):")
print(results_df)

# Imprimir el reporte de clasificación
print("\nReporte de Clasificación para el conjunto TRAIN:\n", classification_report(y_train, y_train_pred))
print("\nReporte de Clasificación para el conjunto TEST:\n", classification_report(y_test, y_test_pred))

# Visualización del árbol de decisión
plt.figure(figsize=(20, 10))
plot_tree(
    model,
    filled=True,
    feature_names=X_train.columns.tolist(),
    class_names=model.classes_.tolist(),
    rounded=True
)
plt.title("Gráfico del Árbol de Decisión (Criterio: Gini, Máx. Profundidad: 10, Min Casos en Nodo Terminal: 45)")
plt.show()
```

## Punto 5: Importancia de predictores

La regla que predice una biena opinión frente el vehículo "Class = Good" es

Baja seguridad, para dos personas, precio de compra muy alto o alto con costo de mantenimiento bajo

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.metrics import confusion_matrix, classification_report

# Configuración del modelo de árbol de decisión con restricción de casos mínimos en nodo terminal
model = DecisionTreeClassifier(criterion="gini", max_depth=10, min_samples_split=2, min_samples_leaf=45, random_state=42)

# Entrenar el modelo en el conjunto TRAIN
model.fit(X_train, y_train)

# --- 1. Gráfico de Importancia de las Variables ---
feature_importances = model.feature_importances_
features = X_train.columns

# Crear un gráfico de barras para las importancias de los predictores
plt.figure(figsize=(10, 6))
plt.barh(features, feature_importances, color="skyblue")
plt.xlabel("Importancia")
plt.title("Importancia de los Predictores")
plt.show()

# --- Función para calcular métricas y generar tablas de clasificación ---
def calculate_metrics(y_true, y_pred):
    confusion = confusion_matrix(y_true, y_pred, labels=["unacc", "acc", "good", "vgood"])
    tn = confusion[0, 0]
    fp = confusion[0, 1:].sum()
    fn = confusion[1:, 0].sum()
    tp = confusion[1:, 1:].sum()
    total = confusion.sum()
    correct_classification = (tp + tn) / total * 100
    fp_percentage = fp / total * 100
    fn_percentage = fn / total * 100
    return correct_classification, fp_percentage, fn_percentage

# Evaluación en el conjunto TRAIN
y_train_pred = model.predict(X_train)
correct_classification_train, fp_train, fn_train = calculate_metrics(y_train, y_train_pred)

# Evaluación en el conjunto TEST
y_test_pred = model.predict(X_test)
correct_classification_test, fp_test, fn_test = calculate_metrics(y_test, y_test_pred)

# Generar la tabla de evaluación
results = {
    "Conjunto": ["Train", "Test"],
    "Clasificación Correcta (%)": [correct_classification_train, correct_classification_test],
    "Falso Positivo (%)": [fp_train, fp_test],
    "Falso Negativo (%)": [fn_train, fn_test]
}
results_df = pd.DataFrame(results)

# Mostrar la tabla comparativa
print("Tabla Comparativa para el Modelo de Árbol de Decisión (Frecuencia Relativa %):")
print(results_df)

# Imprimir el reporte de clasificación
print("\nReporte de Clasificación para el conjunto TRAIN:\n", classification_report(y_train, y_train_pred))
print("\nReporte de Clasificación para el conjunto TEST:\n", classification_report(y_test, y_test_pred))

# --- 2. Tabla de Nodos Terminales ---
# Inicializar listas para los detalles de los nodos terminales
node_ids = []
node_rules = []
class_labels = []
case_counts = []

# Función para recorrer los nodos y extraer información de los nodos terminales
def traverse_tree(node, depth=0, rule=""):
    # Verificar si el nodo es terminal
    if model.tree_.children_left[node] == -1 and model.tree_.children_right[node] == -1:
        # Nodo terminal
        node_ids.append(node)
        node_rules.append(rule)
        
        # Clase más probable en el nodo
        class_counts = model.tree_.value[node][0]
        most_probable_class = model.classes_[np.argmax(class_counts)]
        class_labels.append(most_probable_class)
        
        # Número de casos en el nodo
        case_counts.append(int(model.tree_.n_node_samples[node]))
    else:
        # Nodo no terminal, construir reglas para hijos
        left_child = model.tree_.children_left[node]
        right_child = model.tree_.children_right[node]
        
        # Obtenemos el índice de la característica y el umbral de división
        feature = model.tree_.feature[node]
        threshold = model.tree_.threshold[node]
        
        # Crear regla para el nodo izquierdo
        left_rule = f"{rule} & {features[feature]} <= {threshold:.2f}"
        traverse_tree(left_child, depth + 1, left_rule)
        
        # Crear regla para el nodo derecho
        right_rule = f"{rule} & {features[feature]} > {threshold:.2f}"
        traverse_tree(right_child, depth + 1, right_rule)

# Llamar a la función para comenzar desde la raíz (nodo 0)
traverse_tree(0)

# Crear un DataFrame con los resultados
node_summary_df = pd.DataFrame({
    "Numero del Nodo": node_ids,
    "Regla del Nodo": node_rules,
    "Casos en el Nodo": case_counts,
    "Categoría Predicha": class_labels
})

# Mostrar la tabla de nodos terminales
print("\nResumen de Nodos Terminales:")
print(node_summary_df)

# Visualización del árbol de decisión
plt.figure(figsize=(20, 10))
plot_tree(
    model,
    filled=True,
    feature_names=features,
    class_names=model.classes_.tolist(),
    rounded=True
)
plt.title("Gráfico del Árbol de Decisión (Criterio: Gini, Máx. Profundidad: 10, Min Casos en Nodo Terminal: 45)")
plt.show()
```
